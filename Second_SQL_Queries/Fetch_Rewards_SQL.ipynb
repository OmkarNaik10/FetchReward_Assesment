1) What are the top 5 brands by receipts scanned for most recent month?

```python
conn = sqlite3.connect('FetchRewards.db')
cursor = conn.cursor()

cursor.execute("""
SELECT b.name AS brand_name, COUNT(r.r_id) AS receipts_scanned
FROM rewardReceipt AS rr
JOIN receipt_table AS r ON rr.r_id = r.r_id
JOIN brands_table AS b ON rr.barcode = b.barcode AND rr.brandCode = b.brandCode
WHERE strftime('%Y-%m', datetime(r.dateScanned, 'unixepoch')) = strftime('%Y-%m', 'now')
GROUP BY b.name
ORDER BY receipts_scanned DESC
LIMIT 5;
""")


results = cursor.fetchall()

for row in results:
    print(row)

cursor.close()
conn.close()
```

3) When considering average spend from receipts with 'rewardsReceiptStatus’ of ‘Accepted’ or ‘Rejected’, which is greater?

```python

conn = sqlite3.connect('FetchRewards.db')
cursor = conn.cursor()

cursor.execute("""
SELECT
    rewardsReceiptStatus,
    AVG(totalSpent) AS averageSpend
FROM receipt_table
WHERE rewardsReceiptStatus IN ('Accepted', 'Rejected')
GROUP BY rewardsReceiptStatus
ORDER BY averageSpend DESC;
""")

rows = cursor.fetchall()

for row in rows:
    print("Status:", row[0], "| Average Spend:", row[1])

cursor.close()
conn.close()
```

4) When considering total number of items purchased from receipts with 'rewardsReceiptStatus’ of ‘Accepted’ or ‘Rejected’, which is greater?

```python

conn = sqlite3.connect('FetchRewards.db')
cursor = conn.cursor()


cursor.execute("""
SELECT
    rewardsReceiptStatus,
    SUM(purchasedItemCount) AS totalItemsPurchased
FROM receipt_table
WHERE rewardsReceiptStatus IN ('Accepted', 'Rejected')
GROUP BY rewardsReceiptStatus;
""")

rows = cursor.fetchall()

for row in rows:
    print("Status:", row[0], "| Total Items Purchased:", row[1])

cursor.close()
conn.close()
```

5) Which brand has the most spend among users who were created within the past 6 months?

```python
import sqlite3


conn = sqlite3.connect('FetchRewards.db')
cursor = conn.cursor()


cursor.execute("""
SELECT 
    b.name AS brand_name, 
    SUM(r.totalSpent) AS total_spend
FROM 
    rewardReceipt AS rr
JOIN 
    receipt_table AS r ON rr.r_id = r.r_id
JOIN 
    brands_table AS b ON rr.barcode = b.barcode AND rr.brandCode = b.brandCode
JOIN 
    users_table AS u ON r.userId = u.u_id
WHERE 
    u.createdDate >= strftime('%s', 'now', '-6 months')
GROUP BY 
    b.name
ORDER BY 
    total_spend DESC
LIMIT 1;
""")

result = cursor.fetchone()

if result:
    print("The brand with the most spend among users created within the past 6 months is:", result[0], "with a total spend of", result[1])
else:
    print("No brand spend found for users created within the past 6 months.")

cursor.close()
conn.close()
```

6) Which brand has the most transactions among users who were created within the past 6 months?

```python
import sqlite3

conn = sqlite3.connect('FetchRewards.db')
cursor = conn.cursor()

cursor.execute("""
SELECT 
    b.name AS brand_name, 
    COUNT(rr.r_id) AS transaction_count
FROM 
    rewardReceipt AS rr
JOIN 
    receipt_table AS r ON rr.r_id = r.r_id
JOIN 
    brands_table AS b ON rr.barcode = b.barcode AND rr.brandCode = b.brandCode
JOIN 
    users_table AS u ON r.userId = u.u_id
WHERE 
    u.createdDate >= strftime('%s', 'now', '-6 months')
GROUP BY 
    b.name
ORDER BY 
    transaction_count DESC
LIMIT 1;
""")

result = cursor.fetchone()


if result:
    print("The brand with the most transactions among users created within the past 6 months is:", result[0], "with a transaction count of", result[1])
else:
    print("No transactions found for users created within the past 6 months.")


cursor.close()
conn.close()
```
