# Lets Correct the _id column

```python
# Extract the '$oid' value from each dictionary in the '_id' column
df_brands['_id'] = df_brands['_id'].apply(lambda x: x.get('$oid') if isinstance(x, dict) and '$oid' in x else None)

# Rename '_id' to 'b_id'
df_brands.rename(columns={'_id': 'b_id'}, inplace=True)
```

#  Lets Correct the cpg column

```python
import ast
```

```python
# Ensure the function returns exactly two values, one for cpg_ref and one for cpg_id
def extract_values(row):
    if isinstance(row, dict):
        ref = row.get('$ref', None)
        id_value = row.get('$id', {}).get('$oid', None) if isinstance(row.get('$id', {}), dict) else None
    else:
        ref, id_value = None, None
    return ref, id_value

df_brands['cpg_ref'], df_brands['cpg_id'] = zip(*df_brands['cpg'].apply(extract_values))
```
